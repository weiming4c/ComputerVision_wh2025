# Corner Detection —— Lesson 3.5+

## 核心思想
图片在滑动窗口中微微滑动：
1. 如果任何方向滑动都没有明显变化：flat
2. 如果沿某个方向滑动没有变换，垂直该方向滑动有明显变化：edge
3. 沿任何方向滑动都有变化：corner

## 判断
从lamda判断flat,edge,corner

转化为一个评判量θ

高斯函数窗口下的角点响应函数：
$$
R = g(I_x^2) \cdot g(I_y^2) - [g(I_x I_y)]^2 - \alpha [g(I_x^2) + g(I_y^2)]^2 - t
$$
R是计算得到的角点响应值（Response）。

g()代表高斯滤波操作。$ I_x $ 和 $ I_y $ 分别是图像在 x 和 y 方向的导数。

α是一个经验常数（通常取值在 0.04 到 0.06 之间）。

t是一个阈值，用于最终判断 R是否足够大以被视为角点
